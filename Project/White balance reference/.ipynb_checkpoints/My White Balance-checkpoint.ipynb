```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
import matplotlib.colors as mcolors
```

```python
im = plt.imread('sample2.png')
plt.imshow(im)
```

```python
im_hsv = mcolors.rgb_to_hsv(im[:, :, :3])
fig, (a, b) = plt.subplots(1, 2)
a.imshow(im_hsv[:,:,2])
a.grid()
a.set_xlim(100, 200)
a.set_ylim(200, 150)

b.imshow(im[120:125,160:165])
b.grid()
# b.set_xlim(100, 200)
# b.set_ylim(200, 150)
```
